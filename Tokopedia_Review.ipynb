```python
# Import Library
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import re
import pandas as pd
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)

# Download stopwords jika belum
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# =====================================
# 1. Load Data
# =====================================
# Misal: file berisi kolom 'review'
df = pd.read_csv('kanky.csv')
df = df.dropna(subset=['Ulasan'])  # hilangkan review kosong
df
```

```python
# =====================================
# 2. Preprocessing Text
# =====================================
# Ambil stopwords Bahasa Indonesia dari NLTK
stop_words = set(stopwords.words('indonesian'))

# Tambahkan custom stopwords (singkatan dan kata tidak penting lainnya)
custom_stopwords = {'bgt', 'yg', 'aja', 'dong', 'nih', 'nya', 'deh', 'gk', 'ga', 'tp', 'udh', 'dr','banget', 'it', 'beli', 'kali', 'pol', 'trs', 'bbrp',
                    'gak', 'udah', 'sumpah', 'paa', 'barang', 'sepatu'}
stop_words.update(custom_stopwords)

def preprocess(text):
    text = re.sub(r'[^\w\s]', '', text.lower())
    tokens = text.split()
    tokens = [t for t in tokens if t not in stop_words]
    return ' '.join(tokens)

df['clean_text'] = df['Ulasan'].apply(preprocess)
```

```python
# =====================================
# 3. Sentiment Analysis (sederhana dengan BERT)
# =====================================
vectorizer = CountVectorizer()
dtm = vectorizer.fit_transform(df['clean_text'])
lda = LatentDirichletAllocation(n_components=2, random_state=42)
lda.fit(dtm)
topics = lda.transform(dtm)
df['Topik'] = topics.argmax(axis=1)
```

```python
model_name = "w11wo/indonesian-roberta-base-sentiment-classifier"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

sentiment_pipeline = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
results = sentiment_pipeline(df["Ulasan"].tolist())
df["Sentimen"] = [res["label"] for res in results]
df["Skor_Sentimen"] = [round(res["score"], 2) for res in results]
```
